```python
class Book:
        """This class creates books by taking user inputs regarding all of it's attributes."""
    
    def __init__(self):
        self.isbn_num = int(input('Enter the ISBN number: '))
        self.title = input('Enter the Title of the book: ')
        author = input('Enter the names of authors seperated by comma: ')
        self.authors = list(author.split(','))
        self.publisher = input('Enter the publisher of the book: ')
        self.date = input('Enter the publication date of the book: ')
        self.price = int(input('Enter the price of the book: '))
        
    def __str__(self):
        return f' ISBN Number: {self.isbn_num}, Title: {self.title}, Authors: {self.authors}, Publisher: {self.publisher}, Publication Date: {self.date}, Price: {self.price}'
```

```python
class Library:
        """This class stores all the books and performs all the operations like add a book, 
    remove a book, display all the books,search books by isbn number, search books by author, search books which are lesser tahn a price.
    """
    def __init__(self):
        self.lib = {}
    
    def add(self):
        """Used to add a book to library"""
        book = Book()
        self.lib[book.isbn_num] = [book.title,book.authors,book.publisher,book.date,book.price]
    
    def remove(self,isbn_num):
        """Used to remove a book from the library"""
        del self.lib[isbn_num]    
    
    def display(self):
        """Used to display all the books in the library"""
        print(self.lib)
    
    def search(self,isbn_num):
        """Used to search for book by isbn number"""
        return self.lib.get(isbn_num)
    
    def search_author(self,author):
        """Used to search for books by an author"""
        books = [self.lib[i][0] for i in self.lib if author in self.lib[i][1] ]
        return books

    def search_price(self,price):
        """Used to search for books which are cheaper than a given price"""
        books = [self.lib[i][0] for i in self.lib if self.lib[i][4] < price]
        return books
    
```

```python
def menu():
    """This function gives the available options to the user"""
    print('\n Press 1 for Adding the book')
    print('Press 2 for Deleting the book')
    print('Press 3 for Displaying the books')
    print('Press 4 for Searching the book by isbn number')
    print('Press 5 for Searching the books by an author')
    print('Press 6 for Searching the books lesser than a price')
    print('Press 7 to exit')
    

flag = True 
l = Library()
while(flag == True):
    menu()
    n = int(input('Enter the choice from the menu: '))
    if n==1:
        l.add()
    elif n==2:
        isbn = int(input("Enter the isbn number of the book to be removed: "))
        l.remove(isbn)
    elif n==3:
        l.display()
    elif n==4: 
        isbn = int(input("Enter the isbn number of the book to be searched: "))
        print(l.search(isbn))
    elif n==5:
        author = input("Enter the author of the book to be searched: ")
        print(l.search_author(author))
    elif n==6:
        price = int(input("Enter the price: "))
        print(l.search_price(price))
    elif n==7:
        flag = False
    else:
        print("Enter the correct option!")
    
    
```
